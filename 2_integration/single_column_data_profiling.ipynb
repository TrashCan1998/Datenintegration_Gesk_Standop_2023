```python
import pandas as pd
import numpy as np
import os, itertools
```

```python
# target data
# dir
dir_target = 'target_data_vbn'
# files
file_trips = f'{dir_target}\\trips.txt'
file_stops = f'{dir_target}\\stops.txt'
file_times = f'{dir_target}\\stop_times.txt'
file_agency = f'{dir_target}\\agency.txt'
file_routes = f'{dir_target}\\routes.txt'
file_transfers = f'{dir_target}\\transfers.txt'
```

```python
trips = pd.read_csv(file_trips, low_memory=False)
trips.head(3)
```

```python
trips.describe()
```

```python
trips.isna().sum()
```

```python
trips = trips[['route_id', 'service_id', 'trip_id', 'trip_headsign', 'trip_short_name', 'direction_id']]
trips.head(3)
```

```python
stops = pd.read_csv(file_stops, low_memory=False)
stops.head(3)
```

```python
stops.describe()
```

```python
stops = stops[['stop_id', 'stop_name', 'stop_lat', 'stop_lon']]
stops.head(3)
```

```python
times = pd.read_csv(file_times, low_memory=False)
times.head(3)
```

```python
times.isna().sum()
```

```python
times.drop('stop_headsign', axis=1, inplace=True)
times.head(3)
```

```python
agency = pd.read_csv(file_agency, low_memory=False)
agency.head(3)
```

```python
agency = agency[['agency_id', 'agency_name']]
agency.head(3)
```

```python
routes = pd.read_csv(file_routes, low_memory=False)
routes.head(3)
```

```python
routes.describe()
```

```python
routes.isna().sum()
```

```python
routes = routes[['route_id', 'agency_id', 'route_short_name', 'route_type']]
routes.head(3)
```

```python
transfers = pd.read_csv(file_transfers, low_memory=False)
transfers.head(3)
```

```python
transfers.describe()
```

```python
transfers.isna().sum()
```

```python
transfers = transfers[['from_stop_id', 'to_stop_id', 'transfer_type', 'min_transfer_time']]
transfers.head(3)
```

```python
dir_target = 'target_data_vbn_modified'
trips.to_csv(f'{dir_target}\\trips.csv', index=False)
stops.to_csv(f'{dir_target}\\stops.csv', index=False)
times.to_csv(f'{dir_target}\\stop_times.csv', index=False)
agency.to_csv(f'{dir_target}\\agency.csv', index=False)
routes.to_csv(f'{dir_target}\\routes.csv', index=False)
transfers.to_csv(f'{dir_target}\\transfers.csv', index=False)
```
